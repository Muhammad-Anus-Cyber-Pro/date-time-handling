### **Date time Convert**
---

```python
import numpy as np
import pandas as pd
```

```python
date = pd.read_csv('orders.csv')
time = pd.read_csv('messages.csv')
```

```python
time.head()
```

```python
# Convert string into datetime format
date['date'] = pd.to_datetime(date['date'])
```

```python
date_var = date['date']
date['year'] = date_var.dt.year
date['month'] = date_var.dt.month_name()
date['day'] = date_var.dt.day
```

```python
date.head()
```

```python
# Extract Time
time['date'] = pd.to_datetime(time['date'])
```

```python
time_var = time['date']
time['Hour'] = time_var.dt.hour
time['Min'] = time_var.dt.minute
time['Sec'] = time_var.dt.second
```

```python
a = pd.Timestamp.today()
time.head()
```

### **Date Difference**
---

```python
date['date_diff'] = (pd.Timestamp.today() - date['date']).dt.days
date.head()
```

### **Sort Values**
---

```python
date = date.sort_values('date')
date.head()
```

### **Filtering**
---

```python
date[date['date'] > '2018-07-14'].head()
```

### **Convert Datetime into Nanoseconds**
---

```python
date['date_numeric'] = date['date'].astype('int64')
date.head()
```

### **Relative Days**
---

```python
date['days_since'] = (date['date'] - date['date'].min()).dt.days
date.head(20)
```
